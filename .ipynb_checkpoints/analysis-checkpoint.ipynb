```python
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
from gensim import corpora
import pandas as pd
import numpy as np
from wordcloud import WordCloud
```

```python
# 创建字典
sentences = LineSentence('corpus_without_stopword.txt')
dictionary = corpora.Dictionary(sentences)
```

```python
# 按照词频对token和id排序，展示词频前50
sorted_dictionary = sorted(dictionary.token2id.items(), key=lambda item:dictionary.dfs[dictionary.token2id[item[0]]],reverse=True)
sorted_dictionary[:50]
```

从词频可以看出，小说中出场频率最高的依次是杰洛特，希里，丹德里恩，叶奈法

以前看到这种图总觉得很酷炫，今天顺便看了一下这个词云图的api，发现其实很简单，就顺便加上去了。因为文本的缘故，即使去了停止词，仍然有很多没有什么实际含义的词混在里面，不过还是能看出小说的核心人物

```python
# 使用Word2Vec计算词向量
sentences = LineSentence('corpus.txt')
w2v_model=Word2Vec(sentences,size=100,window=5, min_count=3, workers=4)
vocabulary=w2v_model.wv.vocab
w2v_model.save("model.model")
```

```python
# 计算特定词汇相似度排名前n的词
req_count = 5
for key in w2v_model.wv.similar_by_word('杰洛特', topn =100):
    req_count -= 1
    print (key[0], key[1])
    if req_count == 0:
        break
```

因为“猎魔人”在小说中绝大多数情况下指杰洛特，所以这个两个实体我就没有统一代替。从这个相似度排序，我们发现和杰洛特关系最好的竟然是丹德里恩而不是希里或椰奶。。。
除了小说主要角色以外，和杰洛特关系最近的居然是雷吉斯，也许狩魔猎人的身份让他对非人类更有好感，尤其是充满智慧的高等吸血鬼，对人类的行径更加厌恶。

```python
req_count = 5
for key in w2v_model.wv.similar_by_word('叶奈法', topn =100):
    req_count -= 1
    print (key[0], key[1])
    if req_count == 0:
        break
```

和椰奶关系最近的人竟然是丹德里恩，这简直让利维亚的打桩机本人头上多了一顶大大的绿帽子。。不过可能是算法初始化的影响，加上”猎魔人“这个实体影响了计算，其实结果也不一定准确

```python
req_count = 5
for key in w2v_model.wv.similar_by_word('希里', topn =100):
    if len(key[0])> 2:
        req_count -= 1
        print (key[0], key[1])
        if req_count == 0:
            break
```

```python
print("恩希尔： "+str(w2v_model.wv.similarity('希里','恩希尔')))
print("卡希尔： "+str(w2v_model.wv.similarity('希里','卡希尔')))
print("帕维塔： "+str(w2v_model.wv.similarity('希里','帕维塔')))
```

除了爹妈之外，五人组的米尔瓦是和希里相关度最高的人，作为唯一有资格娶希里（个人观点）的人，卡希尔却都没有出现，可能因为剧情上确实没有太多两人的戏份；大帝和帕维塔公主作为生父生母，跟希里相处的时间缺几乎很少，这也在模型中反映了出来

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['FangSong'] 
mpl.rcParams['axes.unicode_minus'] = False
```

```python
# 提取词向量矩阵
words=list(vocabulary.keys())
vectors = w2v_model[words]
```

```python
# PCA降维
pca = PCA(n_components=2)
dim2_vectors = pca.fit_transform(vectors)
```

```python
# 取出特定词的索引和对应的向量
name_index=[]
names=['杰洛特','叶奈法','希里','恩希尔','大帝','雷吉斯','弗尔泰斯特','卡希尔',
       '特莉丝','邦哈特','菲丽芭','密尔瓦','丹德里恩','卓尔坦','拉多维德','泰莫利亚','瑞达尼亚','科德温','科维尔',
      '尼弗迦德','辛特拉','劳拉朵伦','仙尼德岛','诺维格瑞']
for name in names:
    name_index.append(words.index(name))
name_2d_vectors=dim2_vectors[name_index]
```

```python
# 批标准化
scaler = MinMaxScaler()
scaler.fit(name_2d_vectors)
name_2d_vectors_normalized=scaler.transform(name_2d_vectors)
```

```python
plt.figure(figsize=(18,16))
plt.scatter(name_2d_vectors_normalized[:,0], name_2d_vectors_normalized[:,1])
for i in range(len(names)):
    x = name_2d_vectors_normalized[i][0]
    y = name_2d_vectors_normalized[i][1]
    plt.text(x, y, names[i],fontsize=14,)
plt.title("2D Visualization PCA")
plt.show()
```

pca降维可视化的结果里五人组和希里在左下角被紧紧的圈在一起，但是两个女术士和杰洛特的距离缺远了些，其他相关的实体也被基本圈在了一起

```python
# TSNE降维
tsne = TSNE(n_components=2,init='pca',verbose=1)
dim2_vectors = tsne.fit_transform(vectors)
```

```python
name_2d_vectors=dim2_vectors[name_index]
scaler = MinMaxScaler()
scaler.fit(name_2d_vectors)
name_2d_vectors_normalized=scaler.transform(name_2d_vectors)
plt.figure(figsize=(18,16))
plt.scatter(name_2d_vectors_normalized[:,0], name_2d_vectors_normalized[:,1])
for i in range(len(names)):
    x = name_2d_vectors_normalized[i][0]
    y = name_2d_vectors_normalized[i][1]
    plt.text(x, y, names[i],fontsize=14,)
plt.title("2D Visualization TSNE")
plt.show()
```

TSNE的结果和PCA类似，只不过特征更加收敛了一些

```python
# 3维PCA
pca_3d = PCA(n_components=3)
dim3_vectors = pca_3d.fit_transform(vectors)
name_3d_vectors=dim3_vectors[name_index]
```

```python
scaler = MinMaxScaler()
scaler.fit(name_3d_vectors)
name_3d_vectors_normalized=scaler.transform(name_3d_vectors)
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure(figsize=(18,16)) 
ax = Axes3D(fig) 
ax.scatter(name_3d_vectors_normalized[:,0],name_3d_vectors_normalized[:,1],name_3d_vectors_normalized[:,2]) 
for i in range(len(names)):
    x = name_3d_vectors_normalized[i][0]
    y = name_3d_vectors_normalized[i][1]
    z = name_3d_vectors_normalized[i][2]
    ax.text(x, y, z, names[i],fontsize=16,)
plt.title("3D Visualization PCA",fontsize=16)
```

PCA降到三维的可视化，和二维其实也比较相似
